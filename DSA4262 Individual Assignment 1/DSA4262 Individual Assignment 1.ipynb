##### **Step 1: Removing Unnecessary Columns**

*   **Cases Columns** -> Number of cases is not an accurate metric, different countries have different populations and it is obvious that a country with very high population (ega Chins, India, US) have much higher number of cases than a country with very low population (eg. We will only keep useful columns such as country and Percentage columns in the dataset.

##### **Step 2: Rename Useful Columns To A Readable Format**

For now, the column names DepressiveDisorderByCountryPercentage_2023 is very long and hard to read. We can rename such columns to "Depression Percentage (2023)". This can be applied to all the other year percentage columns from 2019 to 2023. We will also standardize all column names to camel case (eg. country -> Country).

```python
# Renaming the useful columns to a more readable format
depression_rates_country.rename(columns = {'country': 'Country', 
                                           'DepressiveDisorderByCountryPercentage_2023': 'Depression Percentage (2023)',
                                           'DepressiveDisorderByCountryPercentage_2022': 'Depression Percentage (2022)',
                                           'DepressiveDisorderByCountryPercentage_2021': 'Depression Percentage (2021)',
                                           'DepressiveDisorderByCountryPercentage_2020': 'Depression Percentage (2020)',
                                           'DepressiveDisorderByCountryPercentage_2019': 'Depression Percentage (2019)'}, 
                                           inplace = True)

depression_rates_country.head()
```

##### **Step 3: Handling Missing/NaN Values For Useful Columns**

From the dataset result above, we can see that for Syria, the Depression Percentage columns have NaN values, implying that there is missing data for Syria. Other countries listed such as UK, Netherlands and Ukraine have values. There are a couple of possible ways to handle the missing data. 

1) **Drop the rows with missing data** -> To be used if the proportion of NaN values is small (probably less than 10% of the data). 
2) **Imputation using mean/median** -> To be used if the proportion of NaN values is larger (20% or more) such that dropping these rows will affect the accuracy of the depression percentages in subsequent steps.

We first need to determine the proprtion of NaN values for each column. 

```python
# Determining the Number and Proportion of NaN in the dataset
nan_summary = pd.DataFrame({
    "Number Of NaN": depression_rates_country.isna().sum(),
    "Proportion Of NaN": depression_rates_country.isna().sum() / len(depression_rates_country)
})

print(nan_summary)
```

We observe that for the Depression Percentage columns, the proportion of NaN is only 0.089109 (around 9%), which is less than 10% of the data. Hence, we could safely drop the rows with missing data without losing too much information and data. More than 90% of the countries are still present in the dataset. 

```python
# Dropping all rows containing any NaN value
depression_rates_country = depression_rates_country.dropna()
depression_rates_country.head()
```

Here, we observe that the row where the country is Syria has been removed from the dataset due to the presence of NaN values. Now, the first row of the dataset is United Kingdom (originally the second row, after Syria). This shows that the rows containing NaN values have been effectively removed from the dataset.

##### **Step 4: Adding An Additional Column, Continent To Classify The Countries**

This step is necessary as there are too many countries (>180 countries) and imagine creating a bar chart or a line graph where there are such number of categories. Grouping the countries by continent makes much more sense (eg. Asia, Africa, North America, South America, Europe, Oceania) as there will only be 6 categories instead of the initial 180+ categories, making subsequent analysis and visualizations more understandable and more user-friendly. 

We will be using another dataset, Country Continent Groupings, from Our World In Data. This dataset contains the name of the country, the country code, as well as the continent the country is in. 

```python
country_continent = pd.read_excel("Country Continent Groupings.xlsx")
country_continent.head()
```

Similarly, we can clean this dataset by 

1) Selecting the relevant columns - Entity (Country) and World region according to OWID (Continent). Year and Code are not useful columns in our analysis and thus can be removed from the original dataset. 

2) Renaming the useful columns: 
*   Entity -> Country
*   World region according to OWID -> Continent



```python
# Selecting The Relevant Columns - Country and Continent
country_continent = country_continent[['Entity', 'World region according to OWID']]

# Renaming the Relevant Columns
country_continent.rename(columns = {'Entity': 'Country', 
                                    'World region according to OWID': 'Continent'},
                                    inplace = True)

country_continent.head()
```

Afterwards, we can merge the two datasets (country_continent and depression_rates_country) together using the Country column as the column of similarity. We can rely on `pandas pd.merge()` method to combine the two datasets. We will name the merged dataset `depression_rates_continent`.

```python
# Merging The Two Datasets Together (Country Common Column)
depression_rates_continent = pd.merge(depression_rates_country, 
                                      country_continent, 
                                      on = 'Country', 
                                      how = 'left')

# Rearraging The Columns
new_order = ['Country', 'Continent', 'Depression Percentage (2019)', 'Depression Percentage (2020)',
             'Depression Percentage (2021)', 'Depression Percentage (2022)', 'Depression Percentage (2023)']

depression_rates_continent = depression_rates_continent[new_order]
depression_rates_continent.head()
```

Now, the dataset looks much more readable, user-friendly and clean. However, a problem might happen when joining the two datasets. It is possible that a country's name is different for the two datasets that are merged. For example, the `depression_rates_continent` dataset might state **"United States"** while the `country_continent` dataset might state **"United States Of America"**, leading to no joining. We need to verify if such records exist. For these records, the **Continent** column will be represented by **NaN**.

```python
# Finding NaN rows for Continent
rows_with_continent_nan = depression_rates_continent[depression_rates_continent.isna().any(axis=1)]
rows_with_continent_nan
```

Indeed, there is one such row - where country is `Timor-Leste`. Since a **Left Join** is being used, the `depression_rates_country` names the country as "Timor-Leste" but the `country_continent` names the country another way. Upon further investigation, the name stated in `country_continent` is "East Timor". The continent of East Timor/Timor-Leste is supposed to be **Asia**. Hence, we can replace the NaN value with Asia.

```python
# Replace the NaN value in the Continent column to Asia
depression_rates_continent["Continent"] = depression_rates_continent["Continent"].fillna("Asia")

depression_rates_continent[depression_rates_continent["Country"] == "Timor-Leste"]
```

##### **Step 5: Grouping The Countries By Continent and Finding The Average Depression Percentage Rates**

There are now 183 rows in the dataset. If we group by country, there will be too many categories (183 distinct categories) for subsequent analysis and visuslizations. Hence, we will group the countries by continent (6 continents involved) and afterwards, determine the average depression percentage rates for each continent. 

Moreover, we will round each of the average depression rates to 2 decimal places for simplicity, as well as making the visualizations more reader-friendly subsequently.

```python
year_cols = ["Depression Percentage (2019)", "Depression Percentage (2020)",
             "Depression Percentage (2021)", "Depression Percentage (2022)",
             "Depression Percentage (2023)"]

# Grouping the countries by continent and finding the average depression rates
avg_depression_by_continent = (depression_rates_continent.groupby("Continent")[year_cols].mean().reset_index())

# Rounding the average depression rates to 2 decimal places (for simplicity)
avg_depression_by_continent = avg_depression_by_continent.round(2)

avg_depression_by_continent
```

##### **Step 3: Converting Absolute Numbers To Proportions**

For now, the data consists of absolute figures and numbers. This is not ideal as every estate has a different population and it will not be fair when we compare estates using absolute numbers. Some estates such as **Changi** have low population (2530 people only) while other estates such as **Bedok** have high population (289750 people). Hence, it is expected that Bedok will have many more people in each category compared to Changi. However, comparing each category in percentage form will solve the difference in population volume issue. 

```python
# Convert all columns from the 3rd column onwards to numeric
singapore_resident_data.iloc[:, 2:] = singapore_resident_data.iloc[:, 2:].apply(
    pd.to_numeric, errors='coerce')

# Convert the total column (2nd column) to numeric
singapore_resident_data.iloc[:, 1] = pd.to_numeric(
    singapore_resident_data.iloc[:, 1], errors='coerce')

# Divide each age-group column by the total population (row-wise)
for col in singapore_resident_data.columns[2:]:
    singapore_resident_data[col] = (
        singapore_resident_data[col] / singapore_resident_data.iloc[:, 1])

# Multiply all values by 100 to get percentage form
singapore_resident_data.iloc[:, 2:] = singapore_resident_data.iloc[:, 2:] * 100

singapore_resident_data.head(10)
```

##### **What Is DALY?**

Disability-Adjusted Life Year (DALY) is a standard public health metric used to measure the overall burden of disease. One DALY represents one lost year of “healthy” life, combining years lost due to premature death and years lived with disability.

##### **Definition For Percentage Of Total DALYs Caused By Depression:**

The percentage of total DALYs caused by depression refers to the proportion of the total disease burden within an age group that is attributable specifically to depression, relative to all other diseases and health conditions.


To gauge the level of depression among different age groups, we can plot the percentage of total DALYs caused by depression for each respective age groups. The visualization is filled with Blues (color-friendly) as a symbol that this graph represents Males. The higher the percentage of total DALYs caused by depression, the more severe depression contributes to the total overall burden of disease and health conditions. For age groups with higher percentages of total DALYs caused by depression, it indicates that depression is more severe in these age groups and we should look into possible causes of depression more carefully.

#### **Visualization 3 Part 1: Proportion Of 15 To 19 Year Old Females Across Singapore**

To implement the policies (mentioned in the assignment pdf) in Singapore, we first need to zoom in on: 

  1) Which estates in Singapore have the **highest proportion of Females aged 15 to 19?** (This represents the largest group of people with the highest DALY levels related to depression reflected in Visualization 1 Part 3)

  2) Which estates in Singapore have the **highest proportion of Males aged 15 to 19?** (Although the DALY levels related to depression for Teenage Males is lower compared to Teenage Females, it is still significantly high and is the age group with the highest DALY levels related to depression among all Males)

We can acheive this by making use of the data we have imported called **Singapore Resident Population By Age Group And Gender** and our cleaned table **singapore_resident_data** to plot a map of Singapore, with each estate of Singapore having the percentage of 15 to 19 year-old Males and Females respectively. To avoid complexity, we will show two plots of the Singapore map, one for each gender. 

The aim is to **find the estates of Singapore with the highest levels of 15 to 19 year-old females and males** so that the policies can be more implemented in these estates (like more social services support etc.)

Before visualizing, we will need to load the **shapefiles** from **MP14_PLNG_AREA_WEB_PL.shp** file in order to properly draw the Singapore map in the subsequent visualizations. Some data cleaning such as value consistency is required. 

##### **Data Cleaning Steps:**

1) Converting all estate names to camel case for consistency - estate names in the **Singapore Resident Population By Age Group And Gender** file are already in camel case but the estate names in the **MP14_PLNG_AREA_WEB_PL.shp** file are all in capital letter and for effective joining and file reading, we need to ensure consistency. 

2) Remove some estate names from the **MP14_PLNG_AREA_WEB_PL.shp** file. Why? Some estates do not have any residents as they are either industrial estates or forested areas. Names of estates include **Mandai** and **Sungei Kadut**. The rest of the estates have resident populations. The other estates without any residents have been removed during the intial data cleaning process. 

```python
# Loading The ShapeFiles From MP14_PLNG_AREA_WEB_PL.shp
import geopandas as gpd
from matplotlib import cm
from matplotlib.colors import Normalize
import os

# Fix shapefile .shx issue
os.environ["SHAPE_RESTORE_SHX"] = "YES"

# Load shapefile
shapefile_path = "MP14_PLNG_AREA_WEB_PL.shp"
gdf_singapore = gpd.read_file(shapefile_path)

# Normalize estate names to Camel Case
gdf_singapore["PLN_AREA_N"] = gdf_singapore["PLN_AREA_N"].str.title().str.strip()
singapore_resident_data["Estate"] = singapore_resident_data["Estate"].str.title().str.strip()

# Change Mandai to 0 (No one lives in Mandai, jungle area)
singapore_resident_data.loc[singapore_resident_data["Estate"]=="Mandai", "Females Age (15-19)"] = 0
singapore_resident_data.loc[singapore_resident_data["Estate"]=="Mandai", "Males Age (15-19)"] = 0

# Change Sungei Kadut to 0 (No one lives in Sungei Kadut, industrial estate)
singapore_resident_data.loc[singapore_resident_data["Estate"]=="Sungei Kadut", "Females Age (15-19)"] = 0
singapore_resident_data.loc[singapore_resident_data["Estate"]=="Sungei Kadut", "Males Age (15-19)"] = 0

# Merge shapefile with resident data
gdf_merged = gdf_singapore.merge(singapore_resident_data, left_on="PLN_AREA_N", right_on="Estate", how="left")
```

Now, once the data is cleaned and merged successfully, we can start visualizing the proportion of 15 to 19 year-old Females across the various estates in Singapore. We will denote Reds (color-blind friendly) as the color for Females with a darker red indicating a higher percentage. 

```python
# Plotting Visual 3 Part 1 - Proportion Of 15 To 19 Year Old Females Across Singapore

# Column to plot
column_to_plot = "Females Age (15-19)"
values = gdf_merged[column_to_plot].fillna(0)

# Normalize colors with lighter max red
max_val = values.max()
norm = Normalize(vmin=values.min(), vmax=max_val*1.2)
colors = cm.Reds(norm(values))

# Plot map
fig, ax = plt.subplots(figsize=(20, 20))
gdf_merged.plot(color=colors, edgecolor='black', linewidth=0.5, ax=ax)

# Estates to skip labels (dense / overlapping areas)
skip_labels = ["Mandai", "Sungei Kadut", "River Valley", "Singapore River", "Outram", "Downtown Core", "Newton", "Rochor"]

# Add estate names and values
for idx, row in gdf_merged.iterrows():
    val = row[column_to_plot]
    estate_name = row['PLN_AREA_N']

    # Skip NaN and specific dense estates
    if pd.isna(val) or estate_name in skip_labels:
        continue

    x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
    label_color = 'black'
    plt.text(x, y, f"{estate_name}\n{val:.2f}%",
             fontsize=8, ha='center', va='center', color=label_color)

# Add colorbar
sm = plt.cm.ScalarMappable(cmap="Reds", norm=norm)
sm._A = []
cbar = plt.colorbar(sm, ax=ax, shrink=0.5)
cbar.set_label(f"Percentage of Females Aged 15-19 (%)")

plt.title("Percentage of 15-19 Year Old Females Across Singapore Estates (%)", fontsize=24)
plt.axis('off')
plt.tight_layout()
plt.show()
```

From the visualization, we notice that the top 5 estates with the highest proportions of 15 to 19 year-old Females include: 

1) **Pasir Ris** (4.29%)
2) **Choa Chu Kang** (4.10%)
3) **Woodlands** (4.08%) 
4) **Bukit Timah** (3.42%)
5) **Tampines** (3.36%) 

Hence, for these estates with higher proportion of 15 to 19 year-old females, it is recommended that we should prioritse and implement the relevant policies to reduce depression levels/rates of teenage Females in these estates first.

The top 3 estates with the highest proportion of teenage Females and teenage Males are the same, in the same order. 

##### **Possible Reasons Why The Top 5 Estates Are Almost Identical:**

1) **Family-Centric Town Planning**

    Estates like Pasir Ris, Choa Chu Kang, Woodlands, and Tampines were deliberately planned as family-oriented towns: 

    * Large HDB flats (4–5 rooms) suitable for families with children
    * Proximity to schools, childcare centres, parks, and sports facilities

    Since families raise both sons and daughters, the teenage population naturally grows in parallel across genders, leading to similar proportions of teenage males and females.

2) **Cohort Effect from Housing Development Timing**

    Many of these estates experienced major housing development in the late 2000s–2010s. 

    * Couples who moved in during that period now have children aged 15–19
    * Birth cohorts are roughly balanced by gender

    This creates a cohort-driven concentration of teenagers that affects both males and females almost equally, rather than being gender-specific.

#### **Plotting The 2 Graphs Into 1 Visualization**

For the third visualization, we will use subplots - 2 rows and 1 column. The top row will display the **Proportion Of 15 To 19 Year Old Females Across Singapore** and the bottom row will display the **Proportion Of 15 To 19 Year Old Males Across Singapore**. 

```python
# Plotting the 2 graphs into 1 visualization

# Column to plot
columns = ["Females Age (15-19)", "Males Age (15-19)"]

# Females -> Reds, Males -> Blues
colors_map = [cm.Reds, cm.Blues]
titles = ["Percentage of 15-19 Year Old Females Across Singapore Estates (%)",
          "Percentage of 15-19 Year Old Males Across Singapore Estates (%)"]

# Estates to skip labels
skip_labels = ["Mandai", "Sungei Kadut", "River Valley", "Singapore River",
               "Outram", "Downtown Core", "Newton", "Rochor"]

# Create subplots with 2 rows, 1 column
fig, axes = plt.subplots(2, 1, figsize=(20, 20))

for i, col in enumerate(columns):
    ax = axes[i]

    # Values for the column
    values = gdf_merged[col].fillna(0)

    # Lighten top end of the color scale
    max_val = values.max()
    norm = Normalize(vmin=values.min(), vmax=max_val * 1.2)
    colors = colors_map[i](norm(values))

    # Plot the map
    gdf_merged.plot(color=colors, edgecolor='black', linewidth=0.5, ax=ax)

    # Add estate labels
    for idx, row in gdf_merged.iterrows():
        val = row[col]
        estate_name = row['PLN_AREA_N']

        if pd.isna(val) or estate_name in skip_labels:
            continue

        x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
        label_color = 'black'
        ax.text(x, y, f"{estate_name}\n{val:.2f}%", fontsize=8,
                ha='center', va='center', color=label_color)

    # Add colorbar
    sm = plt.cm.ScalarMappable(cmap=colors_map[i], norm=norm)
    sm._A = []
    cbar = plt.colorbar(sm, ax=ax, shrink=0.5)
    cbar.set_label(f"Percentage of {col}")

    # Title and axis off
    ax.set_title(titles[i], fontsize=20)
    ax.axis('off')

plt.tight_layout(h_pad=5)
plt.show()
```
